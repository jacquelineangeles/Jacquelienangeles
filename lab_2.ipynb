Title:– MOLECULAR ACTIVITY AND MEMBRANE TRANSPORT

Purpose: The experiment's purpose is to study passive and active transport mechanisms in cells, with a focus on processes like diffusion, osmosis, and differential permeability. It seeks to understand how materials move in and out of cells without the direct input of cellular energy. Passive transport, driven by factors like Brownian motion and concentration gradients, includes diffusion and osmosis, occurring without ATP expenditure. Active transport, requiring ATP, moves materials against concentration gradients.

Procedure:
A measurement of diffusion through a liquid
1.  fill three Petri dishes with 40 ml of 25°C water.
  2.   Add one crystal of potassium permanganate to each dish, ensuring the same amount is used for all. Record the time.
3. Measure and record the largest colored spot diameter after 5 minutes.
4. Repeat steps 1-3 for water at 5°C and 45°C.

B Measurement of diffusion through agar
1.   Filled agar-filled Petri dishes with two holes. Added two drops of methylene blue to one hole and two drops of potassium permanganate to the other. Record the time and initial spot diameters as time zero.
2.  Measured spot diameters in millimeters every minute for 15 minutes. Calculate average diameters from collective data.
3. Create a graph of average diffusion diameter versus time for both chemicals.
4. Determine diffusion rates for each chemical. Record which, methylene blue or potassium permanganate, has the faster diffusion rate.
5. Consult the Merck Index for molecular formula and structure of methylene blue and potassium permanganate.

D Demonstration of diffusion through a gas


1.   Dampen a cotton ball with 10 drops of 15 M NH4OH and another with 10 drops of 12 M HCl. Place them at opposite ends of an open glass tube. Noted the time.
2.   Observed until white precipitate forms, mark the spot with a pen, and recorded the time.
3. Measured the distance in millimeters from each end to the precipitate mark.
4. Calculated diffusion rates by dividing distance by time. Record which reagent, NH3 or HCl, had the fastest diffusion rate.
5. Explained results based on molecular weight.

f Measurement of osmosis


1.   Securely attach dialysis bags filled with 25% and 50% sucrose solutions to the bottom of two glass tubes immersed in solutions.
2.   Submerge both bags in separate beakers of distilled water, suspending them with ring stand clamps. Check for leaks.
3.  After a five-minute equilibration, mark fluid levels in each tube. Record the time.
4.  Record fluid levels in millimeters every 10 minutes for 50 minutes.
5. If the fluid level reaches the top before 50 minutes, record the time and length from the equilibration line to the top. Calculate the rate in mm/min.
6. Determine osmotic rates for each system. Identify the system with the fastest osmotic rate, 25% or 50% sucrose solution. Provide an explanation for the results.


g Measurement of differential permeability of sugar and starch


1.   Filled a dialysis bag with a 1% starch - 10% glucose solution, ensuring a tight seal.
2. Suspended the bag in a beaker of distilled water using a glass rod.
3. After 15 minutes, tested water for starch and sugar:
a. Tested for starch using Lugol’s solution.
b. Tested for sugar using Benedict’s solution and heat.
4. Repeated water testing at 30, 45, and 60 minutes.
5. Recorded and explained results.



h The effects of tonicity on red blood cells - Demonstration

a. Distilled water (hypotonic)
b. Physiological saline - 0.85% NaCl (isotonic)
c. Saltwater - 2.0% NaCl (hypertonic)

3. Created a wet mount slide for each solution.

4. Examined each slide under the high-dry lens of a compound microscope.

5. Observed:
a. Hemolysis of cells in the hypotonic solution (transparent solution).
